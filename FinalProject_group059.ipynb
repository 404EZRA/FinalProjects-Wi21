Music apps such as Spotify may also use the data to recommend listeners to happier songs to help listeners such that they do not fall victim to the snowball effect of depression caused by listening to sad songs.

# Data Cleaning

After obtaining all of the spotify data from 2018-2020, we still needed to obtain the characteristics of each song such as the danceability, energy, and valence, etc. We used Spotify's API and python library, Spotipy, to obtain the necessary features for every single song in order to analyze our obtained data. 
1. We noticed that some songs appear multiple times per month in the top 200 charts. As a result, we decided to use groupby() to take out the duplicate songs by finding matching song ids and artists. This meant we took out the song title from each observation, which made it easier to analyze the songs we want instead of confusing them with remixes.
2. We then had to use multi indexing in our data to not remove duplicate songs that occurred in different months of the year.
3. The number of streams of a song was also stored as a string, so we had to remove the commas and convert it to a float. We were then able to sum the total number of streams a song had per month.

```python
# Takes scraped and cleaned Top 200 Charts dataset and transforms it to audio features dataset
def create_aud_fts(data):
 
    all_trk_head = pd.DataFrame(aud_ft_dataset(data.head(100)))   # Size must be <= 100 rows
    all_trk_head = all_trk_head.loc[:, ['danceability', 'energy', 'liveness', 'valence', 'tempo', 'id']]
    all_trk_head = all_trk_head.rename(columns = {'id':'Song ID'})
    
    m = 100
    n = 200
    
    while m <= len(data):
        all_trk_next = pd.DataFrame(aud_ft_dataset(data.iloc[m:n]))
        all_trk_next = all_trk_next.loc[:, ['danceability', 'energy', 'liveness', 'valence', 'tempo', 'id']]
        all_trk_next = all_trk_next.rename(columns = {'id':'Song ID'})
        all_trk_head = pd.concat([all_trk_head,all_trk_next])
        
        m += 100; n += 100
        
    all_trk_head.reset_index(drop=True)
        
    return all_trk_head
```

When scraping our Spotify data, we noticed that streams were of type string so we deleted every comma found in the Streams column and turned every string into a float to allow us to manipulate the data as needed. We then created a column with only the year and month for every song as we we were planning to take the averages of each month but we did not want to mix the years together. Finally, there were songs that appeared multiple times every month because we were only able to collect our data as daily streams, so we grouped our data by Artist, Song ID, and Y-M to eliminate any duplicates while not combining data from other months or artists.

After cleaning our covid and Spotify data we needed a way to analyze the averages of valence, energy, and danceability for each month. We could not just group our Spotify data by month and find the average value for each characteristic as this assumes every song carries the same weight. (E.g. a song with a valence value of 0.7 with 100 listens should hold more weight than a song with a valence value of 0.5 and only 10 listens). To solve this issue, we created a function which multiplies the characteristic score of each song by the number of streams and then sums the new value for each song in each month. We then divided this number by the total number of streams for all songs combined in a month and we were left with the average value of a characteristic of our choosing. 

```python
def weighted_scores(df, variable):
    dict_month = {}
    for YM in df['Y-M'].sort_values().unique():
        df_month = df[df['Y-M'] == YM]
        avg_score = np.sum(df_month['Streams']*df_month[variable]) / (sum(df_month['Streams']))
        dict_month[YM] = [avg_score]
        
    weighted_scores_df = pd.DataFrame(dict_month)
    
    return weighted_scores_df
```

# Conclusion & Discussion

In our research, we observed the characteristics of the top 200 weekly Spotify hits from the past three years. We did this by first web scraping from the Spotify top 200 charts and applying the web API, Spotipy, to analyze each song. We combined the weekly data by averaging the streams of each song to create a monthly analysis of the Spotify hits such that we can better observe the songs that had a greater impact over time. We then identified the trends of the characteristics over the past three years and graphed the weighted scores based on the monthly streams which allowed us to equally evaluate the characteristics across 2018, 2019, and 2020. From the tests, we found a positive trend in “Valence'' scores and a negative trend in “Danceability” and “Energy” scores. 

Next, we used the open source library, Prophet, to forecast the 2020 characteristic trends. We created a predictive 2020 model using the 2018 and 2019 data for each characteristic in order to evaluate whether there has been a significant difference in music characteristics. We decided to account for yearly seasonal trends in our predictive model in order to create a better representation of our yearly data and to avoid a purely linear model. From the Mean Absolute Error score alone, we were unable to evaluate whether the predicted models were significantly different from the Actual model. Therefore, we decided to run the Kolmogorov-Smirnov Test to calculate the p-values to determine whether there is a significant difference between the predicted and actual model for each characteristic. Using the Kolmogorov-Smirnov Test, we have identified a significant difference between the predicted and actual model for the “Danceability” and “Energy” characteristics but not for “Valence.” 

In order to identify whether there was a significant difference in characteristics in 2020 which may be attributed to the Pandemic, we chose to analyze the number of Covid cases. Since much of the Pandemic restrictions and traumas are linked to the rise of Covid cases, we believe that the number of Covid cases over the year will be our best metric to analyze the relationships between song characteristics and the Pandemic. From our graphs, we were able to note an inverse relationship between Covid cases and “Danceability,” as well as, “Energy” scores.  As for the “Valence” graph, we identify a peak in the Summer which contrasts the Covid cases, however, we may attribute this peak to potential seasonal consequences. For example, people may be more inclined to listen to happier and upbeat music during Summer break. We were also able to identify a correlation between Covid cases and all three song characteristics using the Pearson R test. Therefore, we can conclude that there is a relationship between Covid cases and the “Danceability'' and “Energy” scores. 

Upon review of our original proposal and feedback, we decided our original question was too broad. We have made changes such that rather than just observing the mood of songs, we are identifying more specific characteristics in relation to the pandemic. In addition, because Professor Ellis suggested finding differences in distribution between a 2020 predicted model and the actual model, we have also expanded our data to include all the months rather than just the timeframe in which the Pandemic began. In doing so, we hoped to observe a continual analysis in trends across the years. In addition we have decided to break out analysis into three main parts: identifying the characteristic trends over the years, identifying whether there is a significant difference in 2020 top hits, and identifying if there is an observable relationship in song characteristics and Covid cases. By doing this three part analysis, we are able to do a more in depth analysis by first establishing that there is a difference in characteristic trends over the years, and then identifying whether the trend has a relationship with the trend in Covid cases.

Our results disproves our hypothesis that all three characteristics were impacted negatively by the pandemic. While we did hypothesize the results about “Danceability” and “Energy” accurately, we did not expect that there wouldn’t be a significant difference in “Valence” score which continued to follow the trend of being more positive over the years. This study allowed us to understand that despite the increase in reported cases of feeling more depressed during the pandemic, we cannot assume that the majority of the American public will turn to sad (low valence and low energy) songs. While there is an increase in popularity for low energy songs, the valence of songs remains relatively the same as before the pandemic. Therefore, people are still inclined to listen to “happy” songs. Our finding does match our expectations that people will turn towards songs that match their mood, as the popular songs in 2020 have a decreased “Danceability” and “Energy” score. This may be due to the fact that people are mostly spending their time at home which leads to a decrease in intensity and activity in their lives (which is reflected in the structure of the songs). 

With a better understanding of how music evolves over time with relationship to large scale historical events such as the pandemic, we are able to better observe the relationship between the sentiment of the public and popular music characteristics. With music as a large factor in the way people express themselves and cope in their lives, it would be helpful for music industries to appeal to the public mass.
