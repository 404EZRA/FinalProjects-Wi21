```python
#We define a function that helps us normalize our data
def normalize_numbers(x):
    if isinstance(x, str):
        val = x.replace(',','')
    else:
        val = x
    try:
        val = float(val)
    except:
        val = np.nan
    return val
```

```python
# obtain the log-scale slope of cases to indicate spread by county
def timeseries_log_slope(series):
    series = series.reset_index(drop=True)
    series = np.log10(series)
    idxarr = series.index.to_numpy()
    A = np.vstack([idxarr, np.ones(len(idxarr))]).T
    m, c = np.linalg.lstsq(A, series,rcond=None)[0]
    DoublingTime = np.log10(2) / m
    return DoublingTime

double_rate_by_county = cases_by_county.iloc[:,4:].apply(lambda x:timeseries_log_slope(x[x>0]), axis=1).to_frame()
double_rate_by_county.columns = ['double_rate']
double_rate_by_county['county_fips'] = cases_by_county['countyFIPS']
double_rate_by_county.drop_duplicates(inplace=True, subset='county_fips', keep="last")
double_rate_by_county = double_rate_by_county.set_index("county_fips").reindex(fips_county_state['county_fips']).reset_index()
double_rate_by_county.head()
```

```python
#We create a function that is able to generate a scatter plot based on the variables passed in.
def scatter_linear(data, x_name, y_name, title = ''):
    data = data[~data.isin([np.nan, np.inf, -np.inf]).any(1)]
    ax = sns.regplot(x=data[x_name], y=data[y_name], color="g")
    plt.title(title)
    # Set x-axis label
    plt.xlabel(x_name)
    # Set y-axis label
    plt.ylabel(y_name)
```

```python
#We also create a function that can generate an OLS report that we can use in our analysis.
def report_OLS(data, x_name, y_name):
    data = data[~data.isin([np.nan, np.inf, -np.inf]).any(1)]
    # print(f'Perform OLS: {y_name}~{x_name}')
    outcome, predictors = patsy.dmatrices(f'{y_name}~{x_name}', data)
    mod = sm.OLS(outcome, predictors)
    res = mod.fit()
    # if res.pvalues[1] < 0.05:
    #     print(res.summary())
    #     print(f'{y_name}~{x_name}')

    corr = data[x_name].corr(data[y_name])
    # print(f'Correlation: {corr}')
    return res
```

```python
# Our goal is to build model that predicts the spread of covid and the death rate given the features
# of a county.
# First, we run OLS to check for variance_inflation_factor of all variables. This shows us if our hand-picked features
# have collinearity issues
# All the feature variables have a variance_inflation_factor less than 5. So the collinearity between our feature
# variables is low.
features = ['rural_percent','hospital_cnt','median_household_income','poverty_percent','population_density']
x = '+'.join(features)
outcome, predictors = patsy.dmatrices(f'cases~{x}', data_by_county)
predictors = sm.add_constant(predictors)
mod = sm.OLS(outcome, predictors)
res = mod.fit()
variables = res.model.exog
vif = [variance_inflation_factor(variables, i) for i in range(variables.shape[1])]
pd.DataFrame({
    'feature': features,
    'variance_inflation_factor': vif[1:]
})
```
