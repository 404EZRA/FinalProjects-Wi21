##### Voter Turnout
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Since the focus is on voter turnout per state, we dropped columns with data on percentage of non citizens, prisoners, probation, parole, total ineligible felons, overseas ineligible votes, and state abbreviation. 

```python
##Cleaning up Turnout

#Chop out unnecessary columns, including ones with null values
turnout = turnout.drop(labels = ['Prison', 'Source', 'Official/Unofficial', 'Probation', 'Parole', 'Total Ineligible Felon', 'Overseas Eligible', 'State Abv', '% Non-citizen', 'Vote for Highest Office (President)'], axis = 1)

#Convert strings in each column to floats
turnout['Total Ballots Counted (Estimate)'] = turnout['Total Ballots Counted (Estimate)'].str.replace(',','').astype(float)
turnout['Voting-Eligible Population (VEP)'] = turnout['Voting-Eligible Population (VEP)'].str.replace(',','').astype(float)
turnout['Voting-Age Population (VAP)'] = turnout['Voting-Age Population (VAP)'].str.replace(',','').astype(float)
turnout['VEP Turnout Rate'] = turnout['VEP Turnout Rate'].str.replace('%','').astype(float)


#Print cleaned dataframe
turnout
```

Total Observations after cleaning: 51

##### Biden v.s. Trump Tweets
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Columns that included impertinent twitter user information such as username, profile description, and join date were dropped. And since the focus is only on states we dropped columns with data on country, continent, city, and state code. Null values and Tweets from non-US states/countries were also dropped.

```python
##Cleaning up Biden Tweets

#Chop out unnecessary columns
biden = biden.drop(labels = ['tweet_id','user_join_date', 'continent', 'country', 'city', 'user_location', 'collected_at', 'state_code', 'tweet', 'source', 'user_id', 'user_name', 'user_screen_name', 'user_description', 'lat', 'long'], axis = 1)

#Drop rows w/ null values
biden = biden.dropna()

#Remove tweets from non-US states
biden = biden[(biden['state'].isin(turnout['State']))]

#Re-index
biden = biden.reset_index()

#Print cleaned dataframe
biden
```

```python
##Cleaning up Trump Tweets

#Chop out unnecessary columns
trump = trump.drop(labels = ['tweet_id','user_join_date', 'continent', 'country', 'city', 'user_location', 'collected_at', 'state_code', 'tweet', 'source', 'user_id', 'user_name', 'user_screen_name', 'user_description', 'lat', 'long'], axis = 1)

#Drop rows w/ null values
trump = trump.dropna()

#Remove tweets from non-US states
trump = trump[(trump['state'].isin(turnout['State']))]

#Re-index
trump = trump.reset_index()

#Print cleaned dataframe
trump
```

```python
##Cleaning State breakdown data

#Chop out unnecessary columns
party_states = party_states.drop(labels = ['county', 'won'], axis = 1)

#Remove parties with candidates other than Trump and Biden and Remove party abbreviations
party_states = party_states[(party_states['party'] == 'DEM') | (party_states['party'] == 'REP')]
party_states = party_states.drop(labels = 'party', axis = 1)

#Group total votes for each candidate in each county by state
party_states = party_states.groupby(['state', 'candidate']).sum().reset_index()

party_states
```

### Analysis

```python
#adding count of tweets by state to turnout dataframe
combined_tweets = pd.concat([biden, trump], ignore_index = True)
combined_tweets = combined_tweets[['created_at', 'likes', 'retweet_count', 'user_followers_count', 'state']]
state_counts = combined_tweets['state'].value_counts()
turnout = turnout.merge(state_counts, left_on = turnout['State'], right_index = True)
turnout = turnout[['State', 'Total Ballots Counted (Estimate)', 'VEP Turnout Rate', 'Voting-Eligible Population (VEP)',
                  'Voting-Age Population (VAP)', 'state']]
turnout = turnout.rename(columns={'state':'tweet_count'})

#corrcoef of count of tweets by state and voter turnout rate
np.corrcoef(turnout['tweet_count'], turnout['VEP Turnout Rate'])
```

With an r value of -0.05 there is a slight negative correlation between tweet count and turnout, such that as tweet count increases the turnout decreases. However, since it is a small value close to 0 there is not a significant correlation present.

```python
#adding avg followers by state to turnout dataframe
state_avg_followers = combined_tweets.groupby(['state'])['user_followers_count'].mean()
turnout = turnout.merge(state_avg_followers, left_on = turnout['State'], right_index = True)
turnout = turnout[['State', 'Total Ballots Counted (Estimate)', 'VEP Turnout Rate', 'Voting-Eligible Population (VEP)',
                  'Voting-Age Population (VAP)', 'tweet_count', 'user_followers_count']]
turnout = turnout.rename(columns={'user_followers_count':'avg_user_followers'})

#corrcoef of avg followers by state and voter turnour rate
np.corrcoef(turnout['avg_user_followers'], turnout['VEP Turnout Rate'])
```

With an r value of -0.04 there is a slight negative correlation between Twitter followers and turnout, such that as Twitter followers increases the turnout decreases. However, since it is a small value close to 0 there is not a significant correlation present.

```python
#adding avg tweet likes by state to turnout dataframe
state_avg_likes = combined_tweets.groupby(['state'])['likes'].mean()
turnout = turnout.merge(state_avg_likes, left_on = turnout['State'], right_index = True)
turnout = turnout[['State', 'Total Ballots Counted (Estimate)', 'VEP Turnout Rate', 'Voting-Eligible Population (VEP)',
                  'Voting-Age Population (VAP)', 'tweet_count', 'avg_user_followers', 'likes']]
turnout = turnout.rename(columns={'likes':'avg_user_likes'})

#corrcoef of avg tweet likes by state and voter turnout rate
np.corrcoef(turnout['avg_user_likes'], turnout['VEP Turnout Rate'])
```

With an r value of -0.12 there is a slight negative correlation between Twitter likes and turnout, such that as Twitter likes increases the turnout decreases. However, since it is a small value close to 0 there is not a significant correlation present, though the r value in this test is closer to 1 compared to correlations between tweet count and followers with voter turnout. 

```python
#adding avg tweet retweet count by state to turnout dataframe
state_avg_retweets = combined_tweets.groupby(['state'])['retweet_count'].mean()
turnout = turnout.merge(state_avg_retweets, left_on = turnout['State'], right_index = True)
turnout = turnout[['State', 'Total Ballots Counted (Estimate)', 'VEP Turnout Rate', 'Voting-Eligible Population (VEP)',
                  'Voting-Age Population (VAP)', 'tweet_count', 'avg_user_followers', 'avg_user_likes', 'retweet_count']]
turnout = turnout.rename(columns={'retweet_count':'avg_user_retweets'})

#corrcoef of avg tweet retweet count by state and voter turnout rate
np.corrcoef(turnout['avg_user_retweets'], turnout['VEP Turnout Rate'])
```

With an r value of -0.12 there is a slight negative correlation between Twitter retweets and turnout, such that as Twitter retweets increases the turnout decreases. However, since it is a small value close to 0 there is not a significant correlation present, though the r value in this test is closer to 1 compared to correlations between tweet count and followers with voter turnout.

With an r value of -0.01 there is a slight negative correlation between voting eligible population and turnout, such that as the voting eligible population increases the turnout decreases. However, since it is a small value close to 0 there is not a significant correlation present, though this being the lowest r value among all the other correlation tests.

```python
#rename VEP Turnout Rate to fit dmatrices syntax
turnout = turnout.rename(columns={'VEP Turnout Rate':'VEPTurnout'})

#linear regression of VEPTurnout and tweet_count
outcome1, predictors1 = patsy.dmatrices('VEPTurnout ~ tweet_count + avg_user_followers + avg_user_likes + avg_user_retweets', data=turnout)
model1 = sm.OLS(outcome1, predictors1)

results1 = model1.fit()

print(results1.summary())
```

The linear regression equation for this graph is Voting Eligible Population=  68.0092 + (-1.133e -05 * tweet count) + (0.001 * average user followers) + (-0.1628 * average user likes) + (0.0773* average user retweets). With this equation when all other variables are equal to 0 the voting eligible population would be 68.0092. If all other variables are left constant for every 1 unit increase in tweet count, you expect to see a 1.133e -05 unit decrease in voting eligible population turnout. If all other variables are left constant for every 1 unit increase in average user followers, you expect to see a 0.001 unit increase in voting eligible population turnout.  If all other variables are left constant for every 1 unit increase in average user likes, you expect to see a 0.1628 unit decrease in voting eligible population turnout.  If all other variables are left constant for every 1 unit increase in average user retweets, you expect to see a 0.0773 unit increase in voting eligible population turnout. These coefficients are minute enough to be considered negligible, indicating that political tweets have little to no effect on voter turnout rate.

The second warning produced by the multivariate regression indicates that some columns within our dataframe are highly correlated, which is to be expected given the nature of social media websites.
