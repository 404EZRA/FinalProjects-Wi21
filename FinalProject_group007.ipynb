# Who Loses the Most Sleep? A Demographic Analysis of Poor Sleepers

# Overview

We took information that the CDC had gathered about which groups in the US get sufficient sleep (defined by the CDC as 7+ hours per night on average) from 2008–2018 and analyzed it to see what demographic trends we could discover around who was losing the most sleep. We used two metrics, slope of the regression line (our way of trying to average the loss of sleep over the 10 year period), and total decline (where we simply compared the values from 2008 and 2018). We discovered that, while some categories did not have any outlier groups in both metrics, the groups losing the most sleep were:  

* Age: 55-64
* Education: Some College
* Household: Single Parent
* Income: Below 100K USD
* Insurance: Public Insurance
* Race: Hispanic

**Note:** We did have to exclude 3 racial groups from the analysis due to excess variability (from which we inferred too small a sample size), as well as excluding sexual orientation as a category altogether for similar reasons.

# Hypothesis


Given the complicated and multi-dimensional nature of demographic splits, our hypothesis is similarly split into several dimensions:
* **Anyone who the medical community critically underserves:** African Americans [1] and women in general [9] face systemic discrimination in the medical system, resulting from unconscious bias reinforced in doctors during their training. It is possible that this has led to worse sleep over time.

* **Demographics with low Socioeconomic Status (SES):** It is well established that those in poverty are more sleep-deprived on average [6]. With income inequality at an all-time high (and still climbing) [11], we hypothesize that this increase will have led to an increase in sleep deprivation.
* **Young people:** With income inequality hitting young people the worst [11], we hypothesize that similar factors will impact this demographic in an exaggerated fashion.
* **People who live in metropolitan areas:** Though industrialization and large cities are by no means new phenomena, the rates of light [12] and noise [13] pollution have continued to rise over time. We hypothesize that these factors may have led to an increase in sleep deprivation in metropolitan areas.

# Dataset(s)

* Dataset Name: Percent of Adults Who Get Sufficient Sleep by Year by State (NHIS_2013_%ofAdultsWhoGetSufficientSleep_byYear_byState.xlsx)  
* Link to the dataset: https://www.healthypeople.gov/2020/data-search/Search-the-Data#objid=5261  
* Number of observations: Undisclosed (see below); pending data request from CDC  

This dataset was collected by the Center of Disease Control (CDC) through their yearly National Health Interview Survey, which collects information on the health of the civilian, noninstitutionalized population of the United States. This particular dataset includes survey information from the years 2008–2018 to calculate the percentage of the adult population that self-report getting a sufficient amount of sleep, which the CDC defines as an average of at least 7 hours of sleep per night. Demographic data is also collected for each participant, allowing the NHIS to break down the adult population into smaller, more detailed categories so the kinds of demographic analyses like the one our question is asking can be performed.  

The biggest weakness in our dataset is that the total population of the participants of the survey are not included in this set, and the results are strictly defined by percentages. We don’t know how many people participated, nor do we have a count of participants per category; we will have to use variability in the data as our metric for deciding which groups may not have had statistically significant sample sizes. While this methodology is far from perfect, it is the best we can do with the existing data. Our best metric of variability is likely to be a combination of yearly standard error and year-to-year variance; too much of each is a likely indicator of an insufficient sample size.  

It is also worth noting that the original spreadsheet only contains state-by-state data for 2013. We will therefore be limited to analyzing national-level data over time.

### Cleaning 5
We still have one problem: thanks to the original csv file having multiple ways of signaling null values, some of the values were turned into strings during the conversion. We can write a function to use with the .apply() method to convert everything to floats. As this only happened with df and not std_err_df, we only need to run this for df.

```python
def change_dtypes(value):
    # These are the three ways the researchers indicated null values. Return np.nan for each
    if value in ['NaN', 'DSU', '---']:
        return np.nan
    else:
        return float(value)

# Apply the changes to each column
for column in df.columns:
    df[column] = df[column].apply(change_dtypes)

# Let's check again
print('Data types in df after reformatting:')
df.dtypes
```

### EDA 1.0.1  

Now that we have an idea of what values missing, we have to assess the reliability of data we do have. We don't have information about how many people were included in each category, so we have to find another way to narrow down which groups might have had sample sizes that make them statistically unreliable for drawing conclusions. Here, we calculate the grand mean (the mean of means) of the standard errors for each demographic category, and use this to gauge a  reasonable StdErr cutoff to determine which groups are "too variable to be reliable". 

```python
# Store average of each column from 'std_err_df' as a list
std_err_means_list = std_err_df.agg('mean').to_list()

# Calculate mean value of list of averages
mean_means = sum(std_err_means_list) / len(std_err_means_list)
print("Grand Mean Standard Error = ", mean_means)
```

### EDA 1.0.2
Given our pooled mean of 1.324, we determine 2 to be our reasonable cutoff. Referencing our cutoff, we identify demographic categories that are highly variable:

```python
# With an aggregate mean StdErr of 1.324, I think 2 is a reasonable cutoff for "too variable to be reliable"

# Store average of each column from 'std_err_df' as a dictionary
std_err_means_dict = std_err_df.agg('mean').to_dict()

# Initialize empty dictionary
std_err_issues = {}

# Iterate StdErr dictionary; add high variability groups to 'issues'
print("Groups with High Variability by Standard Error: ")
for item in std_err_means_dict:
    if std_err_means_dict[item] >= 2:
        std_err_issues[item] = std_err_means_dict[item]
        print('%-20s' '%s' % (item, round(std_err_issues[item], 4)))
```

### EDA 1.1.0
Standard Error isn't our only possible measure of excessive variance. Another possible indicator of small sample size would be too much year-to-year variance, or what we will call the "Yearly Variance Magnitude" (YVM)

```python
def yearly_variance_magnitude(df):
    # Get a list of the columns
    columns = df.columns.tolist()
    
    # Dictionary for storing our outputs
    temp_dict = {}
    
    # Iterate over columns to calculate the yvm for each
    for column in columns:
        year = 2008 # for the index
        total_magnitude = 0 # running total of magnitude
        divisor = 0
        
        # Iterate start year from 2008-2017
        while year < 2018:
            try:
                # If either value is NaN, skip it and don't increase the divisor for our mean
                int(df.loc[year][column])
                int(df.loc[year + 1][column])
                # Take the current year's value and subtract it from the next year's value
                #     Then add the absolute value to total_magnitude
                total_magnitude += abs(df.loc[year + 1][column] - df.loc[year][column])
                divisor += 1
                year += 1
            except:
                year += 1
                
        temp_dict[column] = np.round((total_magnitude / divisor), 3)
    
    # Sort the dictionary
    variance_dict = dict(sorted(temp_dict.items(), key=lambda item: item[1]))
    
    # A few analysis variables
    dict_keys = list(variance_dict.keys())
    variance_min = variance_dict[dict_keys[0]]
    variance_max = variance_dict[dict_keys[-1]]
    variance_mean = np.round((sum(variance_dict.values()) / len(variance_dict)), 3)
    variance_range = np.round((variance_max - variance_min), 3)
    print(' Min variance: ', variance_min, '\n', 'Max variance: ', variance_max, '\n', 
          'Average variance: ', variance_mean, '\n', 'Variance range: ', variance_range)
    
    return variance_dict
```

```python
# Analysis time!
print(' Statistical breakdown of YVM:')
variance_dict = yearly_variance_magnitude(df)

# Code to print them out neatly
print('')
keys = list(variance_dict.keys())
keys_small = keys[:21]
keys_medium = keys[21:42]
keys_large = keys[42:]
for a, b, c in zip(keys_small, keys_medium, keys_large):
    tab_a = '\t'
    tab_b = '\t'
    tab_c = '\t'
    if len(a) < 7:
        tab_a = '\t\t\t'
    elif len(a) < 14:
        tab_a = '\t\t'
    if len(b) < 6:
        tab_b = '\t\t\t'
    elif len(b) < 14:
        tab_b = '\t\t'
    if len(c) < 6:
        tab_c = '\t\t\t'
    elif len(c) < 14:
        tab_c = '\t\t'
    print(a, tab_a, variance_dict[a], '\t', b, tab_b, variance_dict[b], '\t', c, tab_c, variance_dict[c])
```

### Analysis 1.1: Slope of the Regression

There are a couple of ways we can judge which groups are losing the most sleep. First, we can calculate a line of best fit for the data for each group, and compare the slopes of those lines. This is essentially asking, "if this trend continues as it has on average, who can we expect will be worse off later?"

```python
# Calculate the line of best fit to match the line graphs from above
def best_fit_slope(column, df = df):
    x = []
    y = []
    
    for year, item in zip(df.index, df[column]):
        # Filter out nan
        if item > 0:
            x.append(year - 2008)
            y.append(item)
            
    slope, intercept = np.polyfit(x, y, 1)

    return np.round(slope, 3)

# Add in a new column to analysis_df to hold the slope data
analysis_df = analysis_df.assign(slope = analysis_df['demo'])
analysis_df['slope'] = analysis_df['slope'].apply(best_fit_slope)

# Visualize our data
analysis_df.head()
```

### Analysis 1.2: Total Decline

The other, rather simple way to try to answer this question is to simply take the difference of the observed values in 2008 and 2018. For the sake of consistently having negative values to more easily compare this metric to the slope from before, we will subtract 2008 from 2018.

```python
# Calculate the total decline: Where each group started subtracted from where they ended
def total_decline(column, df = df):
    
    data = list(filter(lambda item: item > 0, df[column].tolist()))
    
    return data[-1] - data[0]

# Add in a new column to analysis_df to hold the decline data
analysis_df = analysis_df.assign(decline = analysis_df['demo'])
analysis_df['decline'] = analysis_df['decline'].apply(total_decline)

# Visualize our data
analysis_df.head()
```

### Analysis 1.3.1: Comparison Conclusions

With an R-squared value of .748, we can be sure that our two metrics are at least somewhat strongly related, as we hoped. However, they are not perfectly aligned. Our best results will come from incorporating both of these values.

```python
# 25% value (based on median) for Slope is -0.584
# Let's create a dataframe that drops everything above that value
slope_df = analysis_df.sort_values(by = 'slope', ascending = False)
drop_list = []
for index, row in slope_df.iterrows():
    if row['slope'] > -0.584:
        drop_list.append(index)
        
slope_df = slope_df.drop(drop_list, axis = 0)
```

```python
# 25% value for Decline (based on median) is -4.95
# Let's create a dataframe that drops everything above that value
decline_df = analysis_df.sort_values(by = 'decline', ascending = False)

drop_list = []
for index, row in analysis_df.iterrows():
    if row['decline'] > -4.95:
        drop_list.append(index)
        
decline_df = analysis_df.sort_values(by = 'decline', ascending = False).drop(drop_list, axis = 0)
```

```python
# Find the values in the worst 25% of both slope and decline
both = []
for slope in slope_df.index.tolist():
    if slope in decline_df.index.tolist():
        both.append(slope)

# Create new dataframes with just the values in this list (and 'Total' for comparison)
worst_sleepers_analysis = analysis_df.drop(labels = [i for i in analysis_df.index if 
                                                     (i not in both and i != 'Total')], axis = 0)
worst_sleepers_df = df.drop(labels = [i for i in analysis_df.index if 
                                      (i not in both and i != 'Total')], axis = 1)

# Display the data, grouped by demographic category, then sorted by decline
worst_sleepers_analysis.sort_values(by = ['demo_type', 'decline'])
```

### Analysis 1.4: Deciding on a Primary Metric

Now that we have incorporated both metrics into deciding which groups are generally worst off, we have answered our original question. The groups in `worst_sleepers_df` are the ones who have lost the most sleep over time. However, we notice from the `demo_type` column in `worst_sleepers_analysis` that some categories have more than one group—and that the slope and decline do not necessarily agree for each demographic. If we want to truly answer our question, we should find a single "winner" for each category.

```python
# This function is to help decide which metric to use to calculate "winners"
def slope_or_decline(df = worst_sleepers_analysis, stats_df = analysis_df):
    # Calculate the standard deviation of all slopes and declines
    slope_std = stats_df['slope'].agg('std')
    decline_std = stats_df['decline'].agg('std')
    
    # These are the demographic groups we need to check
    demo_check_list = ['education', 'age_specific', 'income', 'household']
    out_dict = {}
    for demo in demo_check_list:
        # Create a slice of df for only the current demographic group
        temp_df = df[df['demo_type'] == demo]
        
        # Get lists of the slopes and declines, each with 2 values
        slopes = temp_df['slope'].tolist()
        declines = temp_df['decline'].tolist()
        
        # Divide the difference of the slopes and declines by the appropriate std values
        slope_dif = (abs(slopes[0] - slopes[1]) / slope_std)
        decline_dif = (abs(declines[0] - declines[1]) / decline_std)
        
        # Add our decision to the dictionary
        if slope_dif == decline_dif:
            out_dict[demo] = 'Hung Jury'
        elif slope_dif > decline_dif:
            out_dict[demo] = 'Slope'
        else:
            out_dict[demo] = 'Decline'
            
            
    return(out_dict)

slope_or_decline()
```

### Analysis 1.5: Deciding the "Winners"

As it happens, the same demographic "wins" by either metric in the `household` category, so going by this, it seems that Decline should be our primary metric for concluding which demographic gets the title of "worst" in each category. It is also worth noting that the "worst" narrow age group falls within the range of the "worst" broad age group, and has both a larger decline and a steeper slope, so the broad age group has been omitted in favor of the narrow one.

Going by this, the worst sleepers in each demographic category are:

# Ethics & Privacy

Sleep is inherently a personal and private matter, making it a highly difficult subject to collect data on. To protect the privacy of those whose data we use, we only used publicly available and anonymous data sets. Participants were all ages 18 or older. We can also confirm that this data was collected in an ethical fashion by the CDC. While we cannot be sure that they did not accidentally sample in a biased way, we have no reason to suspect that a branch of the government dedicated to the health of the nation had any cause to deliberately bias their data.  

Additionally, we have done our best to avoid making assumptions or generalizations about any demographic groups. Instead, we have sought to find correlations that could shed light onto stressors and the resulting effects they have on the amount of sleep Americans get each night. We also acknowledge that any trends we find in our analysis are far from the only consequences that systemic discrimination has on any demographic group.

# Conclusion & Discussion

Individually, each demographic we identified as having the largest declines in sleep quality can be confirmed by current evidence. In a comparison among Asian, Black/African American, Hispanic, and White race groups, Hispanics report the highest percentage of poor sleep, reported as inability to get enough sleep due to their weekday routine [16]. Meanwhile, in an analysis of adults aged 55–84, 55–64-year-olds were 26 percent more likely to report their sleep as fair to poor and report the highest concentration of sleep problems, at 71% (compared with 64–65%) [17]. Finally, single parents, our category with the most stark decline, are more likely to experience lower sleep quality (42.6 percent) compared with adults in two-parent households (32.7 percent), and adults living without children (31.0 percent) [18]. 

However, it is important we note that our categories are not mutually exclusive; a person can fall into several of the demographic categories with the highest decline in sleep quality. Indeed, there are established relationships between several of our demographics: individuals with public insurance (e.g., Medicaid) by definition earn less than $100k per year; lesser education (e.g., "Some College") is a predictor of lower income; and single parents not only lack a second breadwinner, but must pay the costs of raising a child, further increasing cost of living, and therefore raising their respective poverty threshold. Furthermore, ethnic minorities, such as Hispanics, suffer vast prejudice in the healthcare sector, aptly demonstrated by one Colorado study reporting Hispanics making up 62.1 percent of hospitalizations and 51.2 percent of deaths from the SARS-CoV-2 pandemic [19]. Perhaps the only demographic in our conclusions with no intrinsic relationship to our other categories is age. 

In the present study, we rejected the null hypothesis for two of our five predictions: the medically underserved and low socioeconomic status (SES). However, we failed to reject the null for the remaining three: young people, women, and city-dwellers. The importance of our findings lies largely in future directions; we hope that those doing further research into sleep quality can take our findings into account, placing a special focus on the groups we identified. 

There are a few limitations in the present analysis, however. Firstly, the Center for Disease Control's definition of optimal sleep quality as sleep duration is problematic at best. While sleep quality is indeed contingent on sleep duration, sufficient duration varies from person to person, and such a simplistic metric overlooks factors such as sleep quality, sleep disruption, and biological/psychological constraints on sleep [21]. Secondly, the nature of the CDC’s data as released to the public (percentages and standard error, not individual data points) meant that our conclusions were limited to broad, sweeping judgments. Without data on the level of individuals, a hearty asterisk must be placed on any conclusions we draw, as we cannot conclusively state the interactions that were previously mentioned are, in fact, behind the patterns we see.
