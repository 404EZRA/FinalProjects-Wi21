# Hypothesis


### Hypothesis 1
- Does the expectation of undergraduate students at UCSD increased or decreased during the pandemic compared to the pre-pandemic period?
    - H0(null hypothesis): The remote learning environment does not affect the expectation of undergraduate students. 
    - H1(alternative hypothesis): The remote learning environment affects the expectation of undergraduate students. 

### Hypothesis 2
- Does the average overall academic performance of undergraduate students at UCSD increased or decreased during the pandemic compared to the pre-pandemic period?
    - H0(null hypothesis): The remote learning environment does not affect the average overall academic performance of undergraduate students. 
    - H1(alternative hypothesis): The remote learning environment affects the average overall academic performance of undergraduate students. 

From the hypothesis that we formed, we expect to see that the COVID has negatively affect students' academic performances, mainly by their study hours, average GPA received and grades expected. Since most students do not live in UCSD, they are no longer having access to many resources, such as Geisel Library, some other support from UCSD staff. Lack of face-to-face interaction might also affect students' academic performance since lecture attendance is not longer mandatory. Some students might procrastinate and have trouble focus on the school work with the pressure from COVID-19 itself and other stress.Combined all the factors altogether, we assume that the academic performanace (measured by average GPA and average study hours) and expectation (measured by expected grades)will decrease during the pandemic period.

# Data Cleaning

Parse out course code and course description and extract useful data information from the dataset. We make the study hours to numeric values.

```python
# parse out the course code
cape = cape.assign(course_code = cape['Course'].apply(lambda course: str(course).split('-')[0][:-1]))
cape = cape.assign(department = cape["course_code"].apply(lambda code: str(code).split()[0]))
cape = cape.assign(course_num = cape["course_code"].apply(
    lambda code: str(code).split()[1] 
    if len(str(code).split()) == 2 else code))
# assign the course description
cape = cape.assign(course_description = cape['Course'].apply(
    lambda course: str(course).split('-')[1] if len(str(course).split('-')) == 2 else course))
cape.head()
```

In order to measure the difference of study hours before and during the pandemic, we need data on both studytime_before and studytime_during. Null values are make it difficult to compare the differences and we will remove the null values.

```python
studytime_df = studytime_df.dropna()
print(studytime_df['studytime_during'].unique())
print(studytime_df['studytime_before'].unique())
```

When previewing the data, we found that some data contains units such as \"hours, hour\", we need to git rid of the units and compare only the numeric values.We write out a function to get rid of the units to standardize the study time hours before and during the pandemic.

```python
def removeunit(string):
    string = string.lower()
    string = string.strip()
    
    string = string.replace('hours', '') 
    string = string.replace('hour', '')
    string = string.replace('hrs', '')
        
    string = string.strip()
    return string

studytime_df['studytime_during'] = studytime_df['studytime_during'].apply(removeunit)
studytime_df['studytime_before'] = studytime_df['studytime_before'].apply(removeunit)
studytime_df
studyhour_before = studytime_df['studytime_during']
studytime_df['studytime_during'] = studyhour_before.str.rstrip(' hours hour')
studytime_df.head()
```

Also, we've noticed that some students provided a range of hours instead of a number. For the purpose of analyzing, we will take the average values and use these instead.We transform the datatype into float, and then take the average study hours.

```python
time1 = studytime_df['studytime_during'].apply(lambda studytime_during: str(studytime_during).split('-')[0])
time2 = studytime_df['studytime_during'].apply(lambda studytime_during: str(studytime_during).split('-')[-1])
new_time_during = (time1.astype(float) + time2.astype(float))/ 2

new_time_during
studytime_df['studytime_during'] = new_time_during
```

```python
time1 = studytime_df['studytime_before'].apply(lambda studytime_before: str(studytime_before).split('-')[0])
time2 = studytime_df['studytime_before'].apply(lambda studytime_before: str(studytime_before).split('-')[-1])
new_time_before = (time1.astype(float) + time2.astype(float))/ 2
new_time_before

studytime_df['studytime_before'] = new_time_before
studytime_df.head()
```

```python
def normal_curve(x, mu=0, sigma=1):
    return 1 / np.sqrt(2*np.pi) * np.exp(-(x - mu)**2/(2 * sigma**2))
```

```python
hr_mean = cape['Study Hrs/wk'].mean()
hr_sd = np.std(cape['Study Hrs/wk'])

def plot_clt_curve(sample_size):
    x = np.linspace(0, 10, 8)
    y = normal_curve(x, hr_mean, hr_sd / np.sqrt(sample_size))
    plt.plot(x, y, color='black')
    
plot_clt_curve(500)
```

## Wilcoxon sign-rank test

For analysis, we will be performing Wilcoxon sign-rank test to make inference on relationship between data before and during pandemic. 

We decided to use this test because after plotting our datasets, we fount they do not have a normal distribution. Therefore, we decided to use Wilcoxon sign-rank test to analysis whether there is a significant difference between data before and during the pandemic.

```python
sns.distplot(grade_pandemic["GPA_Expected"])
sns.distplot(grade_pandemic["GPA_Received"])
sns.distplot(grade_pandemic["Study Hrs/wk"])
```

Based on the analysis we have done above, we can see that students’ academic performances, in terms of GPA received, are better. Also, students tend to have a more accurate expectation for their academic performance during the pandemic. Comparing to the analysis done on the survey dataset, the fact that study hours increase during pandemic using remote learning cannot be corroborated, presumably due to our small survey dataset. From our survey result, there is a slight increase in average study hours in remote learning, but the median study hours before and during Covid-19 remains the same, meaning there are some students who increase their study hours during the pandemic, but there is no visible tendency for all students to do so. About a quarter of students think remote learning model at least somewhat negatively affect their academic performance; the majority of students hold a neutral attitude towards remote learning, while the remaining students think their academic performance is positively affected by remote learning. In the case of academic performance that is measured by GPA, there is a visible and statistically proven increase in academic performance in remote learning, which is consistent with the result we found analyzing the cape dataset.

Our hypothesis was that remote learning negatively affected students' learning experiences, which is analyzed from two aspects: students’ expectations and students’ academic performance. Based on our analysis, we found there is a positive influence of remote learning on academic performance and expections where students' gpa received and expected are increasing, rejecting our null hypothesis.

### Discussion

Given that there are only two quarters of data we can gather on students’ learning experiences during the pandemic using remote learning, there are limitations to the observation of the tendencies in our analysis. Only 63 survey observations were collected to triangulate with the results scrapped from CAPE, which is a small dataset and hard to make generalizations from. Gathering more should be the most important step if future research and analysis on this topic are to be done. Besides, the change in school’s grading policies was not taken into consideration when doing the analysis, but it is a confounding variable that can influence students’ academic performance because now students have more time (by the end of the 10th week) to decide whether to take the course as a letter grade or P/NP or even to drop the course.

Additionally, due to the fact that our online survey was a self-reported one, the data was prone to bias. Since remote learning was implemented, students are detached from their normal study and learning environment, away from their peers and instructors. This estrangement may lead to more biases in their self-reported assessment of their academic performance. They have no idea how they, as an individual student, are doing in their classes, but only the overall information provided by teachers (such as midterm mean scores or assignment mean scores). This biased self-assessment may be accounted for the phenomenon presented in our survey data. Some students believe that their academic performance is negatively affected by remote learning, but the average grade received in fact increases during the pandemic.

Although our CAPE analysis suggests that there is a difference in the study hours, average grade expected, and average grade received before and during pandemic and thus rejecting our null hypothesis, we are not able to conclude whether the remote learning model is positively or negatively affecting student’s learning experience. Connecting to our survey results, this project can be improved and more thorough in that we can hypothesize whether there is a positive or negative influence of remote learning and perform one-tailed tests to see the results.
