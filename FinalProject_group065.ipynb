## **Background and Prior Work**

The sedentary lifestyle in America has long plagued its citizens, starting from years before the COVID-19 pandemic. Specifically, the American Fitness and Sports Association states that low-impact activities include walking, jogging, yoga, cycling, or swimming. Unfortunately, many of these activities require special equipment, a safe location, or a comfortable environment, which can be inaccessible to many Americans.

Especially in light of the global COVID-19 pandemic, the sedentary lifestyle has become more   and more prominent. Working from home with little to no outdoor reprieve, especially in dense urban areas and during the cold winter months, it is unsafe and challenging for the average American to engage in traditional exercise. 

Interestingly, the National Health Interview Survey’s conclusion and data showed that exercise frequency varied depending on gender, occupation, and even home state. Researchers found that people working in more white-collar jobs were more likely to get more exercise, and that “high levels of unemployment and disability or poor health in a state were correlated with lower rates of meeting exercise guidelines” (Ducharme, 2018). These findings suggest that certain subgroups are disproportionately disadvantaged when it comes to access to both time and reasonable places to participate in physical activity. As a result, data presented on these topics have to be sensitive to the privilege of something many take for granted: the ability to exercise.

As for our COGS108 project, we are investigating exactly what those privileges may entail. We are primarily focused on finding the link between exercise duration and outdoor temperatures. Some interesting studies that have already been done on parallel topics include a study on the impact of weather on summer and winter exercise behaviors. One particular study published in the Journal of Sport and Health Science found that adverse weather conditions such as rain, snow, or excessive heat, specifically in summer and winter months, delayed an individual’s choice to exercise (Wagner et al, 2019). The study also found that “race, age, and education exacerbate the negative effects of adverse weather conditions on the decision to exercise outdoors” (Wagner et al, 2019.) We are standarizing the values below, which in this project will explicitly be the state variable. The state variable is what we will be using to combine our datasets as we are investigating the relationship between exercise in certain states versus average temperatures of said state. In this case, we need to standardize the values for the datasets by changing each state variable to one standard string (lowercase and fully spelled out) so that it can be combined nicely and follow tidy data principles. In this case, Dataset 1, containing information about exercise location, had various responses in the geographic location category. Some people typed the whole pronoun, spelled incorrectly, or even reported in different languages. Therefore, we standardized these into the full name of the state by editing them to lowercase, as well as stripping the quotes surrounding each state.

```python
# Standardization for data_state
def standardize_state(string):

    # Basic string pre-processing
    string = string.lower()
    string = string.strip()

    # Fix abbreviations
    if string == "ca":
        string = "california"
    if string == "ny":
        string = "new york"
    if string == "wi":
        string = "wisconsin"
    if string == "tn":
        string = "tennessee"
    if string == "md":
        string = "maryland"
    if string == "wa":
        string = "washington"
    if string == "va":
        string = "virginia"
    if string == "sc":
        string = "south carolina"
    if string == "il":
        string = "illinois"
    if string == "ma":
        string = "massachusetts"
    if string == "oh":
        string = "ohio"
    if string == "nj":
        string = "new jersey"
    if string == "mn":
        string = "minnesota"
    if string == "ct":
        string = "connecticut"

    # Fix weird inputs
    string = string.replace("san diego", "california")
    string = string.replace("new york state", "new york") 
    string = string.replace("san francisco bay area", "california") 
    string = string.replace("washington state", "washington") 
    string = string.replace("德州", "texas")
    string = string.replace("密西根州", "michigan")
    string = string.replace("加州", "california")
    string = string.replace("kalifornien", "california")
    string = string.replace("massachussets", "massachusetts")

    # Final strip
    string = string.strip()
    
    return string
```

Looking at the unique values of the percentage of adults that met the physical activity guidelines, we can see that it is presented in the form '% (SE)'. To make this data more easily usable as floats, we want to convert this string into a tuple in the form [%, SE] to easily access the two values. The source summary of this dataset has also noted that some values prefaced with an asterisk '*' signifies that the estimates may be unreliable. For these values, we will set to NaN and drop them from our dataset in a future cell.

```python
# Standardization for data_temp
def standardize_phys_values(string):

    # Check for unreliable data and set to NaN
    if '*' in string:
      string = np.nan
    else:
      # Basic string pre-processing
      string = string.lower()
      string = string.strip()

      # Remove close parenthesis
      string = string.replace(")", "")

      # Tokenize string
      string = [sub.split() for sub in string.split('(')]

      # Construct new tuple
      new_str = []
      for i in string:
          for j in i:
              new_str.append(j)

      string = new_str

      string = tuple(float(s) for s in string)
    
    return string
```

```python
# Apply standardization function to dataset
for column in data_phys.loc[:, data_phys.columns != 'state']:
  data_phys[column] = data_phys[column].apply(standardize_phys_values)

data_phys
```

**Mean Duration and Mean Temperature of Each State**

To better represent our data, we took the the duration of exercise from Dataset 1 and the average temperature readings of each state from Dataset 2 and created a new dataframe that exclusively displayed those two columns. To do this, we created a dictionary to map each state to its proper mean value. This was easily done with the prior data cleaning methods which standardize column names and row values.

```python
# Create a dictionary to organize each state by its mean exercise duration
state_mean_dur_dict = dict()

# Propagate dictionary
for x in data_state['state']:
  state_mean_dur_dict[x] = data_state.loc[data_state['state'] == x]['duration'].mean()

# Create dataframe from dictionary
state_mean_dur = pd.DataFrame(state_mean_dur_dict.items(), columns=['state', 'mean_duration'])
state_mean_dur
```

```python
# Create a dictionary to organize each state by its mean state temperature
state_mean_temp_dict = dict()

# Propagate dictionary
for x in data_temp['state']:
  state_mean_temp_dict[x] = data_temp.loc[data_temp['state'] == x]['average_temp'].mean()

# Create dataframe from dictionary
state_mean_temp = pd.DataFrame(state_mean_temp_dict.items(), columns=['state', 'mean_temp'])
state_mean_temp
```

```python
# Merge the two datasets consisting of the means of interest
merged_df = pd.merge(left=state_mean_temp, right=state_mean_dur, how='left', left_on='state', right_on='state')

# Remove any observations with null values and reset indexing
merged_df = merged_df.dropna()
merged_df.reset_index(drop=True, inplace=True)

merged_df
```

```python
# Append new column holding the two letter abbreviations for each state
merged_df['code'] = merged_df['state'].map(us_state_dict)

merged_df
```

```python
# Set figure size
plt.figure(figsize=(16, 8))

# Plot scatterplot
ax = sns.regplot(x = "mean_temp", y = "mean_duration", data = merged_df)

# Label plots and axes
ax.set_title('Relationship Between Average Temperature and Average Exercise Duration')
plt.xlabel('Average Temperature by State (°C)')
plt.ylabel('Average Exercise Duration by State (Min)')
```

Looking at the scatterplot generated above, we can see that there is a very weak positive correlation between average exercise duration by state and average temperature by state. We can make further observations by using an ordinary least squares model and observe the exact value of the R squared value to see how well fitted the data is with the trend line.

```python
# By convention, the first matrix is the "outcome" or "y" data, and the second is the "predictor" or "x" data
# Try to predict exercise duration from temperature of state
outcome_1, predictors_1 = patsy.dmatrices('mean_duration ~ mean_temp' ,data=merged_df)
mod_1 = sm.OLS(outcome_1, predictors_1)
res_1 = mod_1.fit()
# Print out the summary results of the model fitting
print(res_1.summary())
```

Our null hypothesis is that there is no relationship between temperature and exercise duration. Since the R-squared value of 0.015 is close to 0, the model is not a good fit for our data. Since the p-value of 0.489 is higher than alpha=0.1, temperature does not significantly predict exercise duration. Since p val > alpha, then we fail to reject null hypothesis (aka accept the null hypothesis), which in this case means there is no relationship between temperature and exercise duration.

From this scatterplot measuring the relationship between the percent of adults that meet federal phyical guidelines against the mean temperature by state (°C), it is difficult to discern whether or not there is any strong correlation.The spread of the points is smaller in the middle and larger at the ends so it is not very homoscedastic. We can identify what appears to be a slight negative correlation, in which the lower the mean temperature of the state is, the higher the percent of adults that meet federal physical guidelines. 
However, in order to get a better picture of whether or not this pattern is true and significant, we need to perform an OLS regression on the above data.

```python
# By convention, the first matrix is the "outcome" or "y" data, and the second is the "predictor" or "x" data
# Try to predict exercise duration from temperature of state
percent_pass_first = merged_df2.percent_pass.str[0]
outcome_2, predictors_2 = patsy.dmatrices('percent_pass_first ~ mean_temp' ,data=merged_df2)
mod_2 = sm.OLS(outcome_2, predictors_2)
res_2 = mod_2.fit()
# Print out the summary results of the model fitting
print(res_2.summary())
```

The results of our OLS regression on the relationship between the percent of adults that meet federal phyical guidelines against the mean temperature by state (°C) indicate an R-squared value of 0.221. An R-squared value of 0.221 could be considered meaningful for real world data, as it indicates that about 22% percent variation of one variable is completely explained by the other. This R-squared value is represented by how tightly the data fits the trendline in the above scatterplot. Additionally, if we set our significance level alpha to less than or equal to 0.005,  we can see that the found p-value of 0.005 could be considered significant. The probability of obtaining an effect this extreme with our data is 0.5%. As a result of this OLS regression, we recognize that there is a significant negative correlation between average state temperature and amount of adults in that state who meet federal physical exercise guidelines. As our data analysis findings do not draw a particularly risky conclusion, the possibility of a false positive is not as big of a concern. Therefore, while other factors and variables could have influenced these results, the risks are low enough for us to overlook it and conclude that the lower the state average temperature is, the more likely an adult in that state meets federal guidelines for exercise.

Throughout this project, our team has been interested in the average amount of exercise of any American individual. Our initial question and exploratory interest was on whether or not the type of exercise affected exercise duration. We hypothesized that the higher the average annual temperature of the state was, the more likely an individual from that state will exercise for longer. We came up with this hypothesis because we thought that warmer temperatures meant that it was more likely for individuals to spend time being active outside. But we could only guess as to the reasons why, as our data analysis would not reveal any causal relationships, only correlations. 

Ultimately, our analysis found a weak positive correlation between average temperature and the average exercise duration. Our team decided to explore the statistical results and determined that with a R-squared value of 0.015 , there is little to no correlation between average annual temperature and average exercise duration by state. For our analyses of the percentage of adults from that state that met the federal physical guidelines and the mean temperature of each state, there was a slight negative correlation. The lower the temperatures were, the higher the percentage of adults that met the federal physical guidelines. In our tests, we determined that there was little to no significant difference in the effects of temperature on the duration of exercise. We concluded that our results yielded different results from what we initially hypothesized; however, there were a few confounding variables and factors to consider.

In the end, we decided that temperatures would not be so drastically different over the course of two years as to significantly impact our results; nevertheless, it is still important to acknowledge its potential effects or non-effects. Lastly, we need to consider the type of exercise as well as an individual’s potential access to a gym or home gym. Individuals with a home gym or home exercise routines would be generally unaffected by outdoor temperature. Each of those variables were not accounted for in our statistical analysis.

Surprisingly, we did find that temperature had a greater effect on the state percentage of adults who met the federal physical guidelines. While these findings are not significant enough to suggest anything major, it is important to acknowledge the impact of exercise availability to all populations.
